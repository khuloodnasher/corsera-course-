```python
# counting the number of times the word subject appear in the filee
fname=input('Enter the file name: ')
fhand=open(fname)
count=0
for line in fhand:
    if line.startswith('Subject:'):
        count=count+1
print('There were',count,'Subject lines in', fname)
```

```python
# to avoid wrong file names we will apply try, except, quit functions
# counting the number of times the word subject appear in the filee
fname=input('Enter the file name: ')
# to avoid bad names that might ruit our code we will use try, except, quit as follows
try:
    fhand=open(fname)
except:
    print("The file can not be open ",fname)
    quit()
count=0

for line in fhand:
    if line.startswith('Subject:'):
        count=count+1
print('There were',count,'Subject lines in', fname)

```
